```python
majorroad_df = pd.read_excel("datasets/2013/LAEI2013_MajorRoads_EmissionsbyLink_2013.xlsx")
```

```python
majorroad_df.head()
```

Let's take only the rows with NOx as pollutant

```python
majorroad_NOx_df = majorroad_df[majorroad_df["Pollutant"] == "NOx"]
```

```python
majorroad_NOx_df.shape
```

There are seveal rows with the same GRID_ExactCut_ID

```python
tot = majorroad_NOx_df[majorroad_NOx_df['GRID_ExactCut_ID'] == 24]
tot
```

```python
majorroad_NOx_df.columns
```

Let's drop the columns we do not need

```python
dropped_majorroad_NOx_df = majorroad_NOx_df.drop(['GridId', 'Toid','Location_ExactCut','BoroughName_ExactCut', 'Lts', 'Length (m)', 
                       'Emissions', 'Year','Pollutant', 'Emissions Unit'], axis=1)
```

```python
dropped_majorroad_NOx_df.head()
```

Let's group together the rows with the same GRID_ExactCut_ID and sum the numeric values in the columns for each group

```python
grouped_majorroad_NOx_df = dropped_majorroad_NOx_df.groupby('GRID_ExactCut_ID').sum().reset_index()
```

```python
grouped_majorroad_NOx_df.head()
```

Let's add a colum with the total of the emission from road transportation for each row

```python
grouped_majorroad_NOx_df['NOx_total'] = grouped_majorroad_NOx_df.sum(axis=1)
```

```python
grouped_majorroad_NOx_df.shape
```

Let's import the summary dataset 

```python
emission_NOx_df = pd.read_excel("datasets/2013/LAEI2013_Emissions_Summary-NOx_v1.1.xlsx", sheet_name="NOx by Grid Exact Cut", header=3)
```

```python
emission_NOx_df.head()
```

Let's take only the values for 2013

```python
emission_NOx_2013_df = emission_NOx_df[emission_NOx_df['Year'] == 2013]
```

Let's take GRID_ExactCut_ID, Easting, Northing and total emission 

```python
GRID_ExactCut_ID_EastNort = emission_NOx_2013_df[["GRID_ExactCut_ID","Easting","Northing","Total"]]
```

Now let's join this new df with the previous one to use the info about Easting and Northing for each GRID_ExactCutID and the total emissions

```python
NOx_emission_GRID_df = pd.merge(grouped_majorroad_NOx_df, GRID_ExactCut_ID_EastNort, on=['GRID_ExactCut_ID'], how='inner')
```

```python
grouped_NOx_emission_GRID_df= NOx_emission_GRID_df.groupby('GRID_ExactCut_ID').sum().reset_index()
```

```python
grouped_NOx_emission_GRID_df.head()
```

```python
grouped_NOx_emission_GRID_df.shape
```

```python
grouped_NOx_emission_GRID_df['GRID_ExactCut_ID'].nunique()
```

```python
grouped_NOx_emission_GRID_df.rename(columns={'NOx_total': 'NOx_total_road'}, inplace=True)
```

```python
grouped_NOx_emission_GRID_df.head()
```

In these new dataset we have for each GRID_ExactCut_ID the total of emissions and the total emissions related to road transpost

```python
NOx_emission_conc_GRID_df = pd.merge(grouped_NOx_emission_GRID_df, concentration_NOx_df, on=['Easting', 'Northing'], how='inner')
NOx_emission_conc_GRID_df.head()
```
